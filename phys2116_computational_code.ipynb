```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
#plot planetary radius as a function of equilibrium temperature
#data: temperature (K) and radius (Jupiter radii)
temperature = np.array([1250,1396,2550,1642,1385,1280,2530,1630,2051,1198,1535,1140,2594,1715])
radius = np.array([1.302,1.31,1.512,1.74,1.135,1.13,1.965,1.25,3.91,1.099,1.466,1.279,1.597,1.35])

sort_indices = np.argsort(temperature)
sort_indices = np.argsort(radius)
temperature_sorted = temperature[sort_indices]
radius_sorted = radius[sort_indices]
plt.figure()
plt.scatter(temperature_sorted, radius_sorted)

plt.xlabel('Temperature (K)')
plt.ylabel('Planetary Radius (Jupiter Radii)')
plt.title('Planetary Radius as a Function of Equilibrium Temperature')
plt.show()
```
