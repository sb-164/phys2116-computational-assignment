```python
#calculating the incidence flux of a planet based off of its equilibrium temperature
#s is a substitute for Stefan - Boltzmann constant, W m-2 K-4
#Temperaure is in kelvin (K)
#A_b is bond albedo, amount of light reflected
#incidence flux in W/m^2

s = 5.670374419e-8
T = 2550
A_b = 0.3
flux = (4 * s * T**4) / (1 - A_b)
print(flux)
```
